```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nr_bins = 50
```

```python
import random
goal_x = random.randint(0,nr_bins)
goal_y = random.randint (0, nr_bins)

(goal_x, goal_y)
```

```python
i_df = pd.DataFrame(np.zeros((nr_bins, nr_bins)))

# Create a person:
t=0
x=0
y=0
ped1 = (t, x, y)

i_df[x][y] = 1

diff_x = goal_x - x
diff_y = goal_y - y

for t in range(0, 3):
    last_x =x
    last_y = y
    if diff_x > 0:
        last_x =x
        x += 1
    elif diff_x < 0:
        x -= 1
    if diff_y > 0:
        y += 1
    elif diff_y < 0:
        y -= 1
    i_df[last_x][last_y]=0
    i_df[x][y] = 1

i_df
```

```python
nr_bins = 50
nr_y = 0
nr_x = 0

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111)

for i in range(0, nr_bins):
   ax.axvline(nr_x, color='k', linestyle='-')
   if nr_y < nr_bins:
     ax.axhline(nr_y, color='k', linestyle='-')
     nr_y = nr_y + (1/nr_bins)
   nr_x = nr_x + (1/nr_bins)
```
